```python
import seaborn as sns
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
from  matplotlib.ticker import PercentFormatter

Dataset_1 = pd.read_csv('./Dataset_1.csv')
Dataset_2 = pd.read_csv('./Dataset_2.csv')
Annex = pd.read_csv('./Annex.csv')


# Migration as the Rate of Population

IntMig_Country = pd.merge(Dataset_1,Annex,on='Code')
IntMig_Country = IntMig_Country[(IntMig_Country['Group'] == 'Region') & (IntMig_Country['Gender'] == 'T')       ]
IntMig = IntMig_Country[['Title','Year','Table_2','Table_1','Table_3','Table_6_Est']]

display(IntMig.head(50))

# sns.set_theme(style="whitegrid")
# ax = sns.catplot(data = IntMig, x="Title", y="Table_1", kind="bar",hue = 'Year', height=6, aspect=15/8)
# plt.show()



#titanic = sns.load_dataset("titanic")

IntMig['Table_3'] = IntMig['Table_3'] / 100
g = sns.catplot(x="Year", y="Table_3", col="Title",col_wrap=3,
                 data=IntMig, saturation=20,
                 kind="bar", ci=None, aspect=.6)

#g.fig.set_size_inches(16,8)
#g.fig.subplots_adjust(top=0.81,right=0.86) 



```

```python
IntMig.head()



for i in IntMig.index:

    IntMig.loc[IntMig.index == i,'Ref_Rate'] =  IntMig.loc[IntMig.index == i,"Table_6_Est"] / IntMig.loc[IntMig.index == i,"Table_2"] * 100

IntMig.head()


```

```python

for ax in g.axes.flat:
    ax.yaxis.set_major_formatter(PercentFormatter(1))
    ax.set(ylabel='Migrant / Population', xlabel = 'Year')



plt.show()



IntMig = IntMig[(IntMig['Title'] == 'Africa')]

df = IntMig.copy()
groupedvalues=df.groupby(['Title','Year']).mean().reset_index()

print(df)

pal = sns.color_palette("Greens_d", len(groupedvalues))
rank = groupedvalues["Table_2"].argsort().argsort() 
g=sns.barplot(x='Year',y='Table_3',data=groupedvalues, palette=np.array(pal[::-1])[rank])

for index, row in groupedvalues.iterrows():
    g.text(row.name,row.Table_3, round(row.Table_3 * 100,2), color='black', ha="center")

plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = sns.load_dataset("tips")
groupedvalues=df.groupby('day').sum().reset_index()

print(df)

pal = sns.color_palette("Greens_d", len(groupedvalues))
rank = groupedvalues["total_bill"].argsort().argsort() 
g=sns.barplot(x='day',y='tip',data=groupedvalues, palette=np.array(pal[::-1])[rank])

for index, row in groupedvalues.iterrows():
    g.text(row.name,row.tip, round(row.total_bill,2), color='black', ha="center")

plt.show()
```
